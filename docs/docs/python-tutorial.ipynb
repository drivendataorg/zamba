## Classifying unlabeled videos

Say we want to generate predictions for the videos in a folder called `example_vids` in our home directory.

```python
video_dir = Path('/Users/katewetstone/example_vids')
list(video_dir.iterdir())
```

There are four videos in the folder: `chimp.MP4`, `eleph.MP4`, `blank.MP4`, and `leopard.MP4`. For clarity, here we've named the videos based on what we know each contains. In the real world, your videos will probably have much less helpful names.

```python
# import from zamba
from zamba.models.model_manager import predict_model
from zamba.models.config import PredictConfig
```

```python
help(PredictConfig)
```

#### Simple prediction

Predict the most likely class for each video, and save the results to `example_predictions.csv`

```python
# set up our prediction configuration
predict_config = PredictConfig(data_directory="example_vids/", 
save="example_predictions.csv", 
output_class_names=True)
```

```python
predict_model(predict_config=predict_config)
```
